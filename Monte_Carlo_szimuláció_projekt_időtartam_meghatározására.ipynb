## üìå Egy konkr√©t Monte Carlo szimul√°ci√≥s p√©ld√°t mutatok, amelyben egy projekt id≈ëtartam√°nak becsl√©s√©t v√©gezz√ºk el.

## üì∫ Probl√©mafelvet√©s
Tegyj√ºk fel, hogy egy **IT projektben** t√∂bb f≈ë tev√©kenys√©get kell elv√©gezni, amelyek id≈ëtartama **bizonytalan**. Szeretn√©nk megbecs√ºlni a projekt befejez√©si idej√©t **Monte Carlo szimul√°ci√≥val**.

### **A projekt f√°zisai √©s eloszl√°sai:**
- **Fejleszt√©s (A):** Norm√°lis eloszl√°s√∫, **√°tlag 10 nap, sz√≥r√°s 2 nap**  
- **Tesztel√©s (B):** Exponenci√°lis eloszl√°s√∫, **√°tlag 5 nap**  
- **Telep√≠t√©s (C):** Egyenletes eloszl√°s√∫, **minimum 2 nap, maximum 6 nap**  
- **Tervez√©s (D):** Norm√°lis eloszl√°s√∫, **√°tlag 8 nap, sz√≥r√°s 2 nap**  
- **Dokument√°ci√≥ (E):** Egyenletes eloszl√°s√∫, **minimum 3 nap, maximum 7 nap**  
- **J√≥v√°hagy√°s (F):** Poisson-eloszl√°s√∫, **√°tlag 4 nap**  

### **A projekt teljes id≈ëtartama:**

$$
\text{√ñsszes id≈ë} = A + B + C + D + E + F
$$

Szimul√°ljuk ezt **1000 futtat√°ssal**, √©s n√©zz√ºk meg, mennyi a v√°rhat√≥ befejez√©si id≈ë √©s annak eloszl√°sa.

## üìâ Boxplot √©rtelmez√©se:
A **boxplot** egy gyors √°ttekint√©st ad a projektid≈ëk sz√≥r√°s√°r√≥l √©s eloszl√°s√°r√≥l:

- A **doboz** mutatja az interkvartilis terjedelmet (IQR).  
- A **k√∂z√©ps≈ë vonal** a medi√°nt jelzi.  
- A **"bajuszok"** az adatok minimum √©s maximum √©rt√©keit reprezent√°lj√°k.  

---

## üö® Kiugr√≥ √©rt√©kek (Outliers)
A **boxplot maximuma felett l√©v≈ë pontok** az √∫gynevezett **kiugr√≥ √©rt√©kek**, amelyek **jelent≈ësen elt√©rnek az adatok t√∂bbi r√©sz√©t≈ël**. Ezek olyan projektek, amelyek **j√≥val hosszabb ideig tartottak**, mint az √°tlagos vagy tipikus id≈ëtartamok.

### **Hogyan hat√°rozza meg a boxplot a kiugr√≥ √©rt√©keket?**
A boxplot az **interkvartilis terjedelem (IQR)** alapj√°n hat√°rozza meg a kiugr√≥ √©rt√©keket. Az **IQR** a **25%-os √©s 75%-os percentilisek k√∂z√∂tti k√ºl√∂nbs√©g**:

$$
IQR = Q3 - Q1
$$

A boxplot a kiugr√≥ √©rt√©keket √∫gy defini√°lja, hogy ha egy adatpont k√≠v√ºl esik az al√°bbi tartom√°nyon, akkor **kiugr√≥nak tekintj√ºk**:

$$
\text{Als√≥ hat√°r} = Q1 - 1.5 \times IQR
$$

$$
\text{Fels≈ë hat√°r} = Q3 + 1.5 \times IQR
$$

Ha egy projekt id≈ëtartama enn√©l **nagyobb**, akkor a boxplot felett egy **k√ºl√∂n pontk√©nt** √°br√°zol√≥dik.

---

## üîé Mit jelent ez a Monte Carlo szimul√°ci√≥nkban?
A **fent elhelyezked≈ë pontok azt jelentik, hogy bizonyos szimul√°lt projektek szokatlanul hossz√∫ ideig tartottak**, val√≥sz√≠n≈±leg az√©rt, mert:

- üìå **A tesztel√©si f√°zis (B) exponenci√°lis eloszl√°s√∫**, ami id≈ënk√©nt nagyon nagy √©rt√©keket is vehet fel.  
- üìå **Az √∂sszes id≈ë kisz√°m√≠t√°s√°n√°l egy-egy hosszabb tesztel√©si id≈ë kiemelkedhetett.**  

Ezek a pontok arra utalnak, hogy egyes projektek **extr√©m kock√°zatokkal** is j√°rhatnak. Ha ezek az extr√©m esetek probl√©m√°t jelentenek, akkor √©rdemes:
- **üîª Korl√°tozni a maxim√°lis √©rt√©keket** (pl. az exponenci√°lis eloszl√°st v√°gni egy maximum √©rt√©kn√©l).  
- **üìä Tov√°bbi modellez√©st v√©gezni**, hogy megtudjuk, mi n√∂veli legink√°bb a kock√°zatot.

A **boxplot maximuma felett l√©v≈ë pontok** az √∫gynevezett **kiugr√≥ √©rt√©kek**, amelyek **jelent≈ësen elt√©rnek az adatok t√∂bbi r√©sz√©t≈ël**. Ezek olyan projektek, amelyek **j√≥val hosszabb ideig tartottak**, mint az √°tlagos vagy tipikus id≈ëtartamok.

### **Hogyan hat√°rozza meg a boxplot a kiugr√≥ √©rt√©keket?**
A boxplot az **interkvartilis terjedelem (IQR)** alapj√°n hat√°rozza meg a kiugr√≥ √©rt√©keket. Az **IQR** a **25%-os √©s 75%-os percentilisek k√∂z√∂tti k√ºl√∂nbs√©g**:

$$
IQR = Q3 - Q1
$$

A boxplot a kiugr√≥ √©rt√©keket √∫gy defini√°lja, hogy ha egy adatpont k√≠v√ºl esik az al√°bbi tartom√°nyon, akkor **kiugr√≥nak tekintj√ºk**:

$$
\text{Als√≥ hat√°r} = Q1 - 1.5 \times IQR
$$

$$
\text{Fels≈ë hat√°r} = Q3 + 1.5 \times IQR
$$

Ha egy projekt id≈ëtartama enn√©l **nagyobb**, akkor a boxplot felett egy **k√ºl√∂n pontk√©nt** √°br√°zol√≥dik.

---

# Mit jelent ez a Monte Carlo szimul√°ci√≥nkban?
A **fent elhelyezked≈ë pontok azt jelentik, hogy bizonyos szimul√°lt projektek szokatlanul hossz√∫ ideig tartottak**, val√≥sz√≠n≈±leg az√©rt, mert:

- üìå **A tesztel√©si f√°zis (B) exponenci√°lis eloszl√°s√∫**, ami id≈ënk√©nt nagyon nagy √©rt√©keket is vehet fel.  
- üìå **Az √∂sszes id≈ë kisz√°m√≠t√°s√°n√°l egy-egy hosszabb tesztel√©si id≈ë kiemelkedhetett.**  

Ezek a pontok arra utalnak, hogy egyes projektek **extr√©m kock√°zatokkal** is j√°rhatnak. Ha ezek az extr√©m esetek probl√©m√°t jelentenek, akkor √©rdemes:
- **üîª Korl√°tozni a maxim√°lis √©rt√©keket** (pl. az exponenci√°lis eloszl√°st v√°gni egy maximum √©rt√©kn√©l).  
- **üìä Tov√°bbi modellez√©st v√©gezni**, hogy megtudjuk, mi n√∂veli legink√°bb a kock√°zatot.  

---

# Monte Carlo szimul√°ci√≥hoz az al√°bbi t√≠pus√∫ v√°ltoz√≥k haszn√°lhat√≥k:

### **Folytonos v√°ltoz√≥k:**
- ‚úÖ **Norm√°lis eloszl√°s√∫** (pl. bev√©telek, teljes√≠tm√©nymutat√≥k)  
- ‚úÖ **Egyenletes eloszl√°s√∫** (pl. v√©letlenszer≈± hat√°sokat tartalmaz√≥ adatok)  
- ‚úÖ **Exponenci√°lis eloszl√°s√∫** (pl. esem√©nyek k√∂z√∂tti id≈ë)  
- ‚úÖ **Log-norm√°lis eloszl√°s√∫** (pl. p√©nz√ºgyi hozamok)  

### **Diszkr√©t v√°ltoz√≥k:**
- üé≤ **Binomi√°lis eloszl√°s√∫** (pl. siker/kudarc esetek)  
- üî¢ **Poisson-eloszl√°s√∫** (pl. esem√©nyek sz√°ma egy adott id≈ëszakban)  
- üéØ **Bernoulli-eloszl√°s√∫** (pl. d√∂nt√©si szimul√°ci√≥k)  

---

# Milyen m√©r√©si szint≈± v√°ltoz√≥kat haszn√°lhatsz Monte Carlo szimul√°ci√≥hoz?
- üîπ **Nomin√°lis (n√©vleges) v√°ltoz√≥k**: Nem jellemz≈ë a szimul√°ci√≥ban, kiv√©ve, ha kateg√≥ri√°kat szeretn√©l szimul√°lni (pl. √ºgyf√©lt√≠pusok).  
- üî∏ **Ordin√°lis (sorrendi) v√°ltoz√≥k**: Kev√©sb√© gyakori, de p√©ld√°ul √ºgyf√©l-el√©gedetts√©gi szintek eset√©n alkalmazhat√≥.  
- üîπ **Intervallum v√°ltoz√≥k**: Nagyon gyakori a szimul√°ci√≥s elemz√©sekben (pl. √°rfolyamv√°ltoz√°sok, h≈ëm√©rs√©klet-adatok).  
- üî∏ **Ar√°nyv√°ltoz√≥k**: A leggyakoribb m√©r√©si szint a Monte Carlo szimul√°ci√≥ban (pl. bev√©telek, k√∂lts√©gek, nyeres√©g, id≈ëtartamok).  


# K√≥d

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Monte Carlo szimul√°ci√≥ param√©terei
num_simulations = 1000  # 1000 futtat√°s

# Fejleszt√©s (A): Norm√°lis eloszl√°s (√°tlag=10, sz√≥r√°s=2)
development_times = np.random.normal(loc=10, scale=2, size=num_simulations)

# Tesztel√©s (B): Exponenci√°lis eloszl√°s (√°tlag=5)
testing_times = np.random.exponential(scale=5, size=num_simulations)

# Telep√≠t√©s (C): Egyenletes eloszl√°s (min=2, max=6)
deployment_times = np.random.uniform(low=2, high=6, size=num_simulations)

# Tervez√©s (D): Norm√°lis eloszl√°s (√°tlag=8, sz√≥r√°s=2)
planning_times = np.random.normal(loc=8, scale=2, size=num_simulations)

# Dokument√°ci√≥ (E): Egyenletes eloszl√°s (min=3, max=7)
documentation_times = np.random.uniform(low=3, high=7, size=num_simulations)

# J√≥v√°hagy√°s (F): Poisson-eloszl√°s (√°tlag=4)
approval_times = np.random.poisson(lam=4, size=num_simulations)

# Teljes projektid≈ë kisz√°m√≠t√°sa az √∂sszes f√°zissal
total_project_times = (development_times + testing_times + deployment_times +
                       planning_times + documentation_times + approval_times)

# Kimeneti statisztik√°k
mean_duration = np.mean(total_project_times)
std_duration = np.std(total_project_times)
percentile_5 = np.percentile(total_project_times, 5)
percentile_95 = np.percentile(total_project_times, 95)

# Kimenetek ki√≠r√°sa
print(f"√Åtlagos projekt id≈ëtartam: {mean_duration:.2f} nap")
print(f"Sz√≥r√°s: {std_duration:.2f} nap")
print(f"5%-os percentilis: {percentile_5:.2f} nap")
print(f"95%-os percentilis: {percentile_95:.2f} nap")

# Hisztogram vizualiz√°ci√≥
plt.figure(figsize=(10, 6))
sns.histplot(total_project_times, bins=30, kde=True, color='blue')
plt.axvline(mean_duration, color='red', linestyle='dashed', linewidth=2, label=f'√Åtlag: {mean_duration:.2f} nap')
plt.axvline(percentile_5, color='green', linestyle='dashed', linewidth=2, label=f'5%-os percentilis: {percentile_5:.2f} nap')
plt.axvline(percentile_95, color='orange', linestyle='dashed', linewidth=2, label=f'95%-os percentilis: {percentile_95:.2f} nap')
plt.xlabel("Projekt id≈ëtartam (nap)")
plt.ylabel("Fut√°sok sz√°ma")
plt.title("Monte Carlo szimul√°ci√≥ ‚Äì Friss√≠tett projekt id≈ëtartam eloszl√°sa")
plt.legend()
plt.show()

# ---------------- Tov√°bbi statisztikai sz√°m√≠t√°sok ----------------

# 25%-os √©s 75%-os percentilisek (IQR - Interquartile Range)
percentile_25 = np.percentile(total_project_times, 25)
percentile_75 = np.percentile(total_project_times, 75)
iqr = percentile_75 - percentile_25  # Interkvartilis terjedelem

# Minimum √©s maximum projekt id≈ëtartam
min_duration = np.min(total_project_times)
max_duration = np.max(total_project_times)

# Mod√°lis √©rt√©k (leggyakoribb id≈ëtartam, hisztogram alapj√°n)
hist_values, bin_edges = np.histogram(total_project_times, bins=30)
mode_duration = bin_edges[np.argmax(hist_values)]

# Kimenetek ki√≠r√°sa
print(f"25%-os percentilis (Q1): {percentile_25:.2f} nap")
print(f"75%-os percentilis (Q3): {percentile_75:.2f} nap")
print(f"Interkvartilis terjedelem (IQR): {iqr:.2f} nap")
print(f"Minim√°lis projekt id≈ëtartam: {min_duration:.2f} nap")
print(f"Maxim√°lis projekt id≈ëtartam: {max_duration:.2f} nap")
print(f"Leggyakoribb projekt id≈ëtartam (m√≥dusz): {mode_duration:.2f} nap")

# Boxplot vizualiz√°ci√≥
plt.figure(figsize=(8, 5))
sns.boxplot(x=total_project_times, color='cyan')
plt.xlabel("Projekt id≈ëtartam (nap)")
plt.title("Monte Carlo szimul√°ci√≥ ‚Äì Friss√≠tett projekt id≈ëtartamok boxplot")
plt.show()

# Kiugr√≥ √©rt√©kek azonos√≠t√°sa
outliers = total_project_times[(total_project_times > percentile_75 + 1.5 * iqr)]
print("Kiugr√≥ √©rt√©kek:", outliers)
```
